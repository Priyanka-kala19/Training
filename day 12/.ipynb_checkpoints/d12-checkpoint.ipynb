```python
import numpy as n
import pandas as p
```

```python
data = {'Country':['India','USA','Japan','Russia','Germany'],
        'Year': ['2010','2018','2020','2003','2022'],
        'Continent': ['Asia','NA','Asia','Asia','SA'],
        'Values': [20,3,3,19,27] 
        }
```

```python
dp= p.DataFrame(data)
```

```python
dp
```

```python
p.pivot_table(dp,index=['Continent'],aggfunc='sum')
```

```python
p.pivot_table(dp,index=['Year'],aggfunc='sum')
```

# Concat

```python
people1= {'id': [1,2,3,4],
        'name':['pri','dev','kin','shu'],
         'age':['19','20','15','22'],        
        }

people2= {'id': [11,22,33,44],
        'name':['papa','devang','kinshu','prishu'],
         'age':['17','21','35','20'],        
        }

people3= {'id': [51,52,53,54],
        'name':['nanny','bittu','bharat','tabbi'],
         'age':['89','29','85','66'],        
        }
```

```python
pp=p.DataFrame(people1)
```

```python
pp
```

```python
dd=p.DataFrame(people2)
```

```python
dd
```

```python
kk=p.DataFrame(people3)
```

```python
kk
```

```python
p.concat([pp,dd,kk])
```

```python
p.concat([pp,dd,kk],ignore_index=True)
```

```python
p.concat([pp,dd,kk],axis=1)
```

```python
people4= {'id': [1,2,3,4],
        'name':['pri','dev','kin','shu'],
         'age':['19','20','15','22'],        
        }

people5= {'id': [1,2,3,4],
        'name':['pri','dev','kin','shu'],
         'sales':['1700','2100','3500','2000'],        
        }
```

```python
rr=p.DataFrame(people4)
```

```python
oo=p.DataFrame(people5)
```

```python
p.merge(rr,oo,on=['id','name'])
#in on you have to write that columns name which are same 
```

```python
p.merge(rr,oo,on=['id'])
```
